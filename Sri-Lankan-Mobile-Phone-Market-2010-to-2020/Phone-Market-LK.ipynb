```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import glob
%matplotlib inline
```

```python
path = r'C:\Users\Chinthaka Asiri\Desktop\Python Jupyter\Jupyter 1\Project 7'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df1 = pd.read_csv(filename, index_col=None, header=0)
    li.append(df1)

df_mobile = pd.concat(li, axis=0, ignore_index=True)
df_mobile.head()
```

```python
#Keep the original and work with a copy of data.
mobile=df_mobile.copy()
mobile.fillna(0, inplace=True)
mobile.head()
```

```python
#Understand the dataframe
mobile.shape
```

```python
#Identify duplicates in "Date" Column
mobile.Date.duplicated().sum()
```

```python
#Remove all 4 duplicated columns
mobile.drop_duplicates(subset=['Date'], inplace=True)
mobile.shape
```

```python
mobile['Sum']=mobile.sum(axis=1)
mobile.head()
```

```python
mobile.loc[mobile['Sum'] > 95 , 'Keep the Row'] = True
mobile.loc[mobile['Sum'] <=95 , 'Keep the Row'] = False
mobile['Keep the Row'].value_counts()
```

```python
#Since there are three rows containing lessthan 10% of total, we have to remove them
filter_false = mobile['Keep the Row'] == False
mobile.drop(index=mobile[filter_false].index, inplace=True)
mobile.shape
```

```python
#Double check whether False results have been removed or not
mobile['Keep the Row'].value_counts()
```

```python
#Save the dataframe in to a .csv file and create a new copy of data to proceed with
mobile.to_csv('Filtered-Data.csv',index=0)
mob=mobile.copy()
mob.head()
```

```python
#Clean the dataset - Drop 'Sum' and 'Keep the Row' columns
mob.drop(['Sum','Keep the Row'], axis=1, inplace=True)
mob.head()
```

```python
#Check the dataframe shape for final modifications
mob.shape
```

```python
mob_trans=mob.T 
mt_header=mob_trans.iloc[0]
mob_trans=mob_trans[1:]
mob_trans.columns=mt_header
mob_trans.head()
```

```python
#Save final formated dataframe into CSV file for data visualization 
mob_trans.to_csv('Filtered-Data-T.csv')
```
