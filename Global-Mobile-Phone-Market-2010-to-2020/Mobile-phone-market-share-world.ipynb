```python
import pandas as pd
import glob
%matplotlib inline
```

```python
#Import the CSV file
df_mobile=pd.read_csv('vendor-ww-monthly-201004-202011.csv')
df_mobile.head()
```

```python
#Check for Null values
df_mobile.isnull().sum(axis=0).sum()
```

```python
#Understand the size of the dataframe
df_mobile.shape
```

```python
#Keep a copy of original dataframe and transpose the dataframe
mobile=df_mobile.copy()
mobile.head(2)
```

```python
mobile_t=mobile.T
new_header=mobile_t.iloc[0]
mobile_t=mobile_t[1:]
mobile_t.columns=new_header
mobile_t.head()
```

```python
#Save final formated dataframe into CSV file for data visualization 
mobile_t.to_csv('Filtered-Data-T.csv')
```
